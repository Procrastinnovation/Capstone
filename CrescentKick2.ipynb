```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import pandas
import math

df1 = pandas.read_csv('csv/20200321_122217_Accelerometer_crescent2.csv') #Front Kick

HostTime = df1['HostTimestamp']
HostTime -= HostTime[0]
HostTime /= 1000

HostTime = np.array(HostTime)
df1 = df1[df1.HostTimestamp > 0]
HostTime = HostTime[HostTime > 0]

acc1 = df1['X (mg)']**2 + df1['Y (mg)']**2 + df1['Z (mg)']**2
acc1 = np.sqrt(acc1)

accX = df1['X (mg)']
accY = df1['Y (mg)']
accZ = df1['Z (mg)']

acc1=np.array(acc1)/1000

plt.plot(HostTime,acc1,'r-')
plt.xlabel('time [s]')
plt.ylabel('acceleration [g]')
plt.show()

# create figure and axis objects with subplots()
fig,ax=plt.subplots()
ax.plot(HostTime, acc1, marker="o")
ax.set_xlabel("time")
ax.set_ylabel("speed")
plt.show()

df2 = pandas.read_csv('csv/20200321_122217_Gyroscope_crescent2.csv')

HostTime = df2['HostTimestamp']
HostTime -= HostTime[0]
HostTime /= 1000

HostTime = np.array(HostTime)
df2 = df2[df2.HostTimestamp > 0]
HostTime = HostTime[HostTime > 0]

GyrX = df2['X (dps)']
GyrY = df2['Y (dps)']
GyrZ = df2['Z (dps)']


plt.plot(HostTime,GyrX,'r-')
plt.xlabel('time [s]')
plt.ylabel('X_Ang_vel [degrees/second]')
plt.show()

plt.plot(HostTime,GyrY,'r-')
plt.xlabel('time [s]')
plt.ylabel('Y_Ang_vel [degrees/second]')
plt.show()

plt.plot(HostTime,GyrZ,'r-')
plt.xlabel('time [s]')
plt.ylabel('Z_Ang_vel [degrees/second]')
plt.show()
```
